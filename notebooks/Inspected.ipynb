### Question 2: *Clean the “Inspection Outcome” variable. It is a good practice to merge categories with few observations in an “Other” category. Code this implementation without “hardcoding”.*

My strategy here is to set a threshold percentage, and then set any INSPECTIONOUTCOME value with a count lower than that treshold to Other.

```python
import numpy as np
threshold = len(merged) * 0.01
outcomes = merged["INSPECTIONOUTCOME"].value_counts()
to_be_changed = []
for outcome, count in outcomes.items():
    if count < threshold:
        to_be_changed.append(outcome)
merged.loc[merged["INSPECTIONOUTCOME"].isin(to_be_changed), "INSPECTIONOUTCOME"] = "Other"
merged["INSPECTIONOUTCOME"].value_counts()
```

### Conclusion: *The INSPECTIONOUTCOME column is now cleaned up. I arbitrarily chose 1% as my treshold, but I'd like to know if there's a rule of thumb for that kind of thing.*
