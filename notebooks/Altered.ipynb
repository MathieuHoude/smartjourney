# Analysis of the *Altered* Dataset
### Question 1: *Merge the data : We want to keep information about the previously merged dataset even if elevators were not altered.*

```python
import pandas as pd
license_installed = pd.read_csv("../data/processed/License-Installed.csv")
altered = pd.read_json("../data/raw/altered.json")
altered.columns = altered.columns.str.upper().str.replace(' ', '')
merged = pd.merge(license_installed, altered, how='outer', on="ELEVATINGDEVICESNUMBER")
merged
```

### Conclusion: *The dataframe from the previous section and the altered dataset have now been merged.*
- - - 

### Question 2: *Are there elevators with more than 5 alterations ? If positive, show all rows about those elevators.*

```python
high_alteration_occurences = merged.groupby(['ELEVATINGDEVICESNUMBER', 'LOCATIONOFDEVICE']).filter(lambda x: len(x) > 5)
high_alteration_occurences.groupby(['ELEVATINGDEVICESNUMBER', 'LOCATIONOFDEVICE']).count()
```

### Conclusion: *The dataframe for elevators with more than 5 alterations has been created.*
